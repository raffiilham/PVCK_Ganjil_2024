### **Transformasi Linier Brightness**

Formula:g(x,y)=f(x,y)+b

g(x,y) adalah nilai pixel setelah transformasi, f(x,y)adalah nilai pixel asli, b adalah nilai brightness

```python
print(' mengubah tingkat kecerahan citra ')
print('----------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan:'))
except ValueError:
  print('Error, not a number')

img = cv.imread('/content/drive/MyDrive/PCVK/images/female.png')
# the variable name was changed from original to img
brightness_image = np.zeros(img.shape, img.dtype)

#akses per pixel
for y in range(img.shape[0]):
  for x in range(img.shape[1]):
    for c in range(img.shape[2]):
      brightness_image[y,x,c] = np.clip(img[y,x,c] + brightness, 0, 255)

#cara simple tanpa for loop
#brightness_image = cv.convertScalesAbs(img, beta=brightness)

final_frame = cv.hconcat([img, brightness_image])
cv2_imshow(final_frame)
```

1. Implementasikan inverse citra pada Google Colaboratory menggunakan formula yang terdapat pada bagian Ulasan Teori, sehingga menghasilkan keluaran seperti berikut:

```python
img = cv.imread('/content/drive/MyDrive/PCVK/images/ktp.jpeg')
# the variable name was changed from original to img
inverse_citra = np.zeros(img.shape, img.dtype)

#akses per pixel
for y in range(img.shape[0]):
  for x in range(img.shape[1]):
    for c in range(img.shape[2]):
      inverse_citra[y,x,c] = np.clip (255 - img[y,x,c] , 0, 255) # Added the a_min argument which is 0

final_frame = cv.hconcat([img, inverse_citra])
cv2_imshow(final_frame)
```

2. Implementasikan transformasi contrast pada Google Colaboratory menggunakan formula yang terdapat pada bagian Ulasan Teori untuk kontras, sehingga menghasilkan keluaran seperti berikut:

```python
print(' mengubah kontras dan tingkat kecerahan citra ')
print('----------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan [-255 - 255]:'))
  contrast = int(input('Masukkan kontras [1.0 - 3.0]:'))
except ValueError:
  print('Error, not a number')

img = cv.imread('/content/drive/MyDrive/PCVK/images/female.png')
# the variable name was changed from original to img
brightness_image = np.zeros(img.shape, img.dtype)

#akses per pixel
for y in range(img.shape[0]):
  for x in range(img.shape[1]):
    for c in range(img.shape[2]):
      brightness_image[y,x,c] = np.clip(contrast * img[y,x,c] + brightness, 0, 255)

#cara simple tanpa for loop
#brightness_image = cv.convertScalesAbs(img, beta=brightness)

final_frame = cv.hconcat([img, brightness_image])
cv2_imshow(final_frame)
```

3. Implementasikan transformasi logarithmic brightness pada Google Colaboratory menggunakan formula yang terdapat pada bagian Ulasan Teori untuk transformasi log, sehingga menghasilkan keluaran seperti berikut:

```python
print(' mengubah tingkat kecerahan citra dengan Transformasi Log')
print('----------------------------------')
try:
  brightness = int(input('Masukkan nilai kecerahan:'))
except ValueError:
  print('Error, not a number')

img = cv.imread('/content/drive/MyDrive/PCVK/images/female.png')
# the variable name was changed from original to img

# Logarithmic brightness transformation
I_max = 255
brightness_image = np.zeros(img.shape, img.dtype)

for y in range(img.shape[0]):
    for x in range(img.shape[1]):
        for c in range(img.shape[2]):
            brightness_image[y, x, c] = np.log(1 + img[y, x, c]) / np.log(1 + I_max) * 255

#cara simple tanpa for loop
#brightness_image = cv.convertScalesAbs(img, beta=brightness)

final_frame = cv.hconcat([img, brightness_image])
cv2_imshow(final_frame)
```
